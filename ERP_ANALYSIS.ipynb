# Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import random
import copy
```

# Preparing Random Signals 

```python
trails=50
sample_rate=100
baseline=0.5
sec_for_each_trials=2
signal_time=(sec_for_each_trials+baseline)*sample_rate
classes=[]
for i in range(2):
    random_signals=[]
    for j in range(trails):
        random_signals.append(np.random.uniform(0, 2.5,int(signal_time)))
    classes.append(random_signals)
```

```python
plt.plot(classes[0][0],color = 'blue')
plt.title('random signal 1')
plt.grid(True, which='both')
plt.show()
```

```python
plt.plot(classes[1][0],color = 'red')
plt.title('random signal 2')
plt.grid(True, which='both')
plt.show()
```

```python
old_classes = copy.deepcopy(classes)
```

# Baseline Correlation

```python
for i in range(2):
    for j in range(trails):
        baseline_mean=(np.array(classes[i][j][:50]).mean())
        classes[i][j]=classes[i][j][-200:]
        for k in classes[i][j]:
            k-=baseline_mean
```

```python
fig, (ax1, ax2) = plt.subplots(2)

ax1.plot(old_classes[0][0],color = 'blue')

ax2.plot(classes[0][0],color = 'red')
plt.show()
```

# ERP

```python
ERP=[]
for i in range(2):
    each_sample=[]
    for j in range(200):
        sum=0
        for k in range(50):
            sum+=classes[i][k][j]
        each_sample.append(sum/50.0)
    ERP.append(each_sample)
```

```python
fig, (ax1, ax2) = plt.subplots(2)

ax1.plot(ERP[0],color = 'blue')

ax2.plot(ERP[1],color = 'red')
plt.show()
```
